```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection, metrics #neighbors, preprocessing, 
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold


from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
df.columns = ["sepal length", "sepal width", "petal length", "petal width", "class"]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.set()
sns.pairplot(df, hue='class',height=3)
plt.show()
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(df.iloc[:,:-1], 
                                                    df.iloc[:,-1], test_size=0.25)
```

```python
df.iloc[:,-1].unique()
```

```python
y_train[y_train.iloc[0:] == 'Iris-versicolor'] = 'r'
y_train[y_train.iloc[0:] == 'Iris-virginica'] = 'g'
y_train[y_train.iloc[0:] == 'Iris-setosa'] = 'b'
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X_train.iloc[:,0:1], X_train.iloc[:,3:4], c=y_train[:], s=350, cmap='viridis')
plt.title('Training data')
plt.show()
```

```python
y_test[y_test.iloc[0:] == 'Iris-versicolor'] = 'r'
y_test[y_test.iloc[0:] == 'Iris-virginica'] = 'g'
y_test[y_test.iloc[0:] == 'Iris-setosa'] = 'b'

plt.figure(figsize=(10,10))
plt.scatter(X_test.iloc[:,0:1], X_test.iloc[:,3:4], c=y_test[:], s=350, cmap='viridis')
plt.title('Test data')
plt.show()
```

```python
y_train[y_train.iloc[0:] == 'r'] = 'Iris-versicolor'
y_train[y_train.iloc[0:] == 'g'] = 'Iris-virginica'
y_train[y_train.iloc[0:] == 'b'] = 'Iris-setosa'

y_test[y_test.iloc[0:] == 'r'] = 'Iris-versicolor'
y_test[y_test.iloc[0:] == 'g'] = 'Iris-virginica'
y_test[y_test.iloc[0:] == 'b'] = 'Iris-setosa'
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier(n_estimators=100, random_state=101)
```

```python
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
rfc_cv_score = cross_val_score(clf, X, y, cv=10)
```

```python
print("All AUC Scores")
print(rfc_cv_score)
```

```python
print("Mean AUC Score")
print("Mean AUC Score - Random Forest Classification: ", rfc_cv_score.mean())
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix")
cm = confusion_matrix(y_test, y_pred)
cm
```

# KNN

```python
from sklearn import neighbors
```

```python
knn = neighbors.KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
accuracy = knn.score(X_test, y_test)
```

```python
knn.get_params()
```

```python
y_pred=knn.predict(X_test)
```

```python
print("KNN Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("=== KNN Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred))
```

```python
print("=== KNN Classification Report ===")
print(classification_report(y_test, y_pred))
```

```python
knn_cv_score = cross_val_score(knn, X, y, cv=10)
```

```python
print("=== KNN All AUC Scores ===")
print(knn_cv_score)
```

```python
print("=== Mean AUC Score ===")
print("Mean AUC Score - KNN: ", knn_cv_score.mean())
```

# Naïve Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train) 
```

```python
y_pred= nb.predict(X_test)
```

```python
print("Naive Bayes Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
print("=== Naive Bayes Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred))
```

```python
print("=== Naive Bayes Classification Report ===")
print(classification_report(y_test, y_pred))
```

```python
nb_cv_score = cross_val_score(nb, X, y, cv=10)
```

```python
print("=== Naive Bayes All AUC Scores ===")
print(nb_cv_score)
```

```python
print("=== Mean AUC Score ===")
print("Mean AUC Score - Naive Bayes: ", nb_cv_score.mean())
```
